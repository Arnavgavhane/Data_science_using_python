```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x=np.linspace(0,10)
y=np.linspace(0,8)

plt.plot(x,y,color='red')
plt.title("sine wave")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=np.linspace(0,50,10)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label="sine",color="yellow",linestyle="-")
plt.plot(x,y2,label="cosine",color="blue",linestyle="-.")
plt.title("sine wave and cosine wave")
plt.xlabel('year')
plt.ylabel('price')
plt.legend()
plt.grid(True)
plt.show()
```

## multiple sub plots

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

fig,axs=plt.subplots(2)
axs[0].plot(x,y1,color="green")
axs[0].set_title("sine")
axs[1].plot(x,y2,color="red")
axs[1].set_title("cosine")
plt.tight_layout
plt.show()
```

## bar chart

```python
fruits=["apple","orange","banana","kiwi"]
prices=[100,80,40,150]

plt.bar(fruits,prices,color="#000080")
plt.title("prices of fruits")
plt.xlabel("fruits")
plt.ylabel("prices")
plt.show()
```

## piechart

```python
subject=['physics','chemistry','maths','biology','history','electronics','hindi']
marks=[89,78,65,76,67,76,92]
plt.pie(marks,labels=subject,autopct='%1.2f%%',startangle=180)
plt.axis("equal")
plt.title("subject wise marks")
plt.show()
```

```python


marks = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']

plt.pie(marks, labels=labels)
# No axis("equal") here!
plt.title("Without axis('equal')")
plt.show()
```

```python
plt.pie(marks, labels=labels)
plt.axis("equal")
plt.title("With axis('equal')")
plt.show()
```

```python
movies=["endgame","joker","ford vs ferrari","the martian","the hangover"]
ratings=[8.4,8.4,8.1,8.0,7.7]

plt.bar(movies,ratings,color="#000080")
plt.title("movie ratings")
plt.xlabel("movies")
plt.ylabel("ratings")
plt.show()
```

## histogram

```python
data = np.random.randn(10)
plt.hist(data,color="orange",edgecolor="green",bins=6)
plt.title("histogram")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.show()
```

## political pie chart

```python
party_name=['BJP','INC','CPI(M)','CPI']
members=[202000000,55000000,1020000,3130000]
plt.pie(members,labels=party_name)
plt.title("political party members")
plt.show()
```

## histogram with matplotlib

```python
plt.figure(figsize=(8,5))
plt.hist(members,color='green',edgecolor='red',bins=5)
plt.title('histogram')
plt.grid(True)
plt.xlabel("members")
plt.ylabel('party')
plt.show()
```

## Histogram with seaborn

```python
df=pd.DataFrame({"members":members})
print(df)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['members'],bins=5,kde=True,color='green')
plt.title('histogram with kde')
plt.xlabel('members')
plt.ylabel('density')
plt.show()
```
