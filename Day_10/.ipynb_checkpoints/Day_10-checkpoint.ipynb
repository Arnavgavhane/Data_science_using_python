## Titanic EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
print(titanic.head(10))
```

```python
print(titanic.tail(10))
```

## shape of data

```python
print(titanic.shape)
```

```python
print(titanic.describe)
```

```python
print(titanic.dtypes)
```

### to print the sum of null values

```python
print('null values\n',titanic.isnull().sum())
```

#### summary stat

```python
print('summary\n',titanic.describe(include='all'))
```

#### correlation matrix

```python
print('correlation\n',titanic.select_dtypes(include='number').corr())
```

#### drop columns with too many missing values

```python
titanic=titanic.drop(['deck'],axis=1)
print(titanic)
```

#### converting survived column from 1-0 to survived-not survived


```python
titanic['survived']= titanic['survived'].map({0:'not-survived',1:'survived'})
print(titanic)
```

#### count of survived vs not survived

```python
sns.countplot(x='survived',data=titanic)
plt.title('survival count')
plt.xlabel('survived (0=no,1=yes)')
plt.ylabel('count')
plt.show()
```

```python
sns.countplot(x='sex',data=titanic)
plt.title('survival count per age')
plt.xlabel('survived (0=no,1=yes)')
plt.ylabel('count')
plt.show()
```

```python
sns.countplot(x='sex',hue='survived',data=titanic)
plt.title('survival count by sex')
plt.xlabel('survived male and female')
plt.ylabel('count')
plt.show()
```

### survival by class

```python
sns.countplot(x='pclass',hue='survived',data=titanic)
plt.title('survived by passenger class')
plt.show()
```

#### town wise survival

```python
sns.countplot(x='embark_town',hue='survived',data=titanic)
plt.title('town wise survival')
plt.show()
```

### age distribution

```python
plt.hist(titanic['age'].dropna(),bins=30,color='blue',edgecolor='red')
plt.title('age distribution')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

### corelation heat map

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('correlation heat map')
plt.show()
```
