# K- Means Clustering

What is K-Means?
- K-Means groups data into clusters based on similarity 
- you can tell the algorithm how many groups (clusters) you want (K)
- it finds the centre of those groups and assigns data point to the nearest one
- example : you have mixed fruit sizes => k-means seperates them into small,medium and large groups without being told in advance

what is clustering 
- clustering means `grouping similar things together`
- you dont tell the computer the groups, it finds them based on patterns
- real life example : imagine you have a basket of mixed fruits (apple,oranges,bananas) with no labels. the computer looks at size, colour, shape and groups similar fruits together 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['species']= data.target
df.head()
```

```python
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df)
plt.title('petal length vs petal width')
plt.show()
```

```python
X=df[['petal length (cm)','petal width (cm)']]
kmeans= KMeans(n_clusters = 3,random_state=42)
kmeans.fit(X)
```

```python
df['cluster']=kmeans.labels_
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df,hue='cluster',palette='deep')
plt.title('petal length vs petal width')
```

# plot cluster centers

```python
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],s = 200,c='black',marker='X',label='Centres')
plt.title('kmeans clustering of samples(petal features)')
plt.legend()
plt.show()
```
