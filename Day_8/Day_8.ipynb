## box plot for each feature

```python
plt.figure(figsize=(12,6))
for i,col in enumerate(df.columns[:-1]):
    plt.subplot(2,2,i+1)
    sns.boxplot(x='species',y=col,data=df)
    plt.title(f"{col} by species")
plt.tight_layout()
plt.show()
```
